Matplottlib is most Popular data visualization library. It provides the control over ever aspect of figure.
It is built on Numpy library.

```python
import matplotlib.pyplot as ppy
import numpy as np
%matplotlib inline
```

```python
''' There are two ways to create a plot using Matplotlib
    1) Functional Approch
    2) Object Oriented Approch
'''
```

```python
# Using Functional Approch
x=np.arange(0,11)
ppy.plot(x,x**5)
ppy.show()
```

```python
# Now we can change the colour of our plot line, name axis and add title to plot
ppy.plot(x,x**2,'r-')
ppy.xlabel("X Label")
ppy.ylabel("y LABLE")
ppy.title("Title")
```

```python
# We can create subplots on single Canvas (using functional approch)
ppy.subplot(2,2,1)
ppy.plot(x,x**3)
ppy.subplot(2,2,2)
ppy.plot(x,x**5,'r-')
ppy.subplot(2,2,3)
ppy.plot(x,x//2,'g-')
ppy.subplot(2,2,4)
ppy.plot(x,x+5,'y-')
# To avoid overlapping
ppy.tight_layout()
```

```python
#Using Object Oriented Approch
fig=ppy.figure()  #figure object is created
axes=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.3,0.3])
axes.set_xlabel('X label')
axes.set_ylabel('Y lable')
axes.set_title("TITLE")
axes.plot(x,x**2,'-g')
axes2.plot(x**2,x,'-y')
```

```python
# Making subplots using object oriented method
fix,axes=ppy.subplots(nrows=3,ncols=1)
for i in axes:
    i.plot(x,x//2,'-c')
axes[1].plot(x*22,x)
axes[0].set_title('1st subplot')
axes[1].set_title('2st subplot')
axes[2].set_title('3st subplot')
ppy.tight_layout()
```

```python
''' Figure Size aspect ration and DPI '''
fig = ppy.figure(figsize=(3,2),dpi=100,facecolor='g')
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
```

```python
fig,((ax1,ax2),(ax3,ax4),(ax5,ax6)) = ppy.subplots(nrows=3,ncols=2,figsize=(5,5),facecolor='y')
ax1.plot(x,x**2)
ax2.plot(x**2,x)
ax3.plot(x,x//2)
ax4.plot(x//2,x)
ax5.plot(x,x+2)
ax6.plot(x,5-x)
```

```python
# How to save a figure
''' We can save a figure in various formats using Matplotlib function'''
fig.savefig('MY_Figure.png',dpi=500)
```

```python
# Legends
''' 
Legends are used to Describe plot, Which plot gives what information.
'''
fig = ppy.figure(dpi=100,facecolor='c')
ax = fig.add_axes([0,0,1,1])
ax.plot(x+1,x**2,label='A')
ax.plot(x+2,x**2,label='B')
ax.plot(x+3,x**2,label='C')
ax.plot(x+4,x**2,label='D')
ax.plot(x+5,x**2,label='E')
ax.legend(loc=0)
```

```python
# customising our plot
fig = ppy.figure(figsize=(5,3),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x^2,color='orange',linewidth=3,alpha=1,linestyle='-.',marker='o',markersize=15,markerfacecolor='yellow'
        ,markeredgecolor='red')
```
